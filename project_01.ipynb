```python
import pandas as pd 
import os 
import json
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='latin1')
df.head(5)
```

```python
df.info()
```

```python
df.describe(include = "all")
```

```python
Missing_values = df.isnull().sum()
```

```python
Missing_values
```

```python
df_filtered = df[df["SALES"]>1000]
```

```python
df_filtered
```

```python
folder_name = "cleaned_data"
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
```

```python
df_filtered.to_csv(f"{folder_name}/filtered_sales.csv",index = False)
```

```python
summary = {
    "Total_Rows" : len(df),
    "Missing_values" : df.isnull().sum().to_dict(),
    "Filtered_Rows(sales > 1000)" : len(df_filtered)
}
```

```python
with open("cleaned_data/summary.json","w") as f:
    json.dump(summary,f,indent = 5)
```
